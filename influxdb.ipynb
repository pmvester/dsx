```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
host = "169.46.20.39"
port = 8086
user = "root"
password = "root"
dbname = "k34db"
```

```python
from influxdb import DataFrameClient
client = DataFrameClient(host, port, user, password, dbname)
```

```python
import collections
pdict = client.query("select * from pressure")['pressure']['pressure'].to_dict()
spdict = collections.OrderedDict(sorted(pdict.items()))
```

```python
pvalues = map(lambda x : x / 100.0, spdict.values())
plt.boxplot(pvalues, showmeans=True)
plt.show()
```

```python
plt.hist(pvalues)
plt.show()
```

```python
plt.plot(spdict.keys(), map(lambda x : x / 100.0, spdict.values()))
plt.show()
```
